```python
# Colab typically has ffmpeg preinstalled. If not, uncomment:
# !apt-get update -qq && apt-get install -y ffmpeg

import sys
from pathlib import Path

ROOT = Path('.').resolve()
if str(ROOT) not in sys.path:
    sys.path.append(str(ROOT))

print('Root:', ROOT)
```

```python
# MODE 2: upload script.txt
uploaded = {}
try:
    from google.colab import files
    uploaded = files.upload()
except Exception as e:
    print('Upload skipped:', e)

script_path = None
if 'script.txt' in uploaded:
    in_path = Path('data/input/script.txt')
    in_path.parent.mkdir(parents=True, exist_ok=True)
    in_path.write_bytes(uploaded['script.txt'])
    script_path = in_path
    print('Saved:', script_path)
```

```python
# MODE 1: topic (used if no script.txt uploaded)
topic = '3 habits that make you mentally strong'
```

```python
from IPython.display import Video, display

from src.pipeline.content_pipeline import run_pipeline
from src.providers.simple_script_provider import SimpleScriptProvider
from src.utils.logging_utils import setup_logging

setup_logging('INFO')

result = run_pipeline(
    topic=None if script_path else topic,
    script_file=script_path,
    script_provider=SimpleScriptProvider(),
)

print('Video:', result.video_path)
print('Hashtags:', ' '.join(result.hashtags))
display(Video(str(result.video_path), embed=True))
```
